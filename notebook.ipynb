# Data

We aggregated a few datasets found online into a dataset for **Parkinsons** and a dataset for **Parkinsons, ALS, and cerebral palsy**. 

We cleaned them and ended up with 2 versions of each : one with the NaN rows **removed**, and one where we replaced them with the **medians**. 

Let's do some exploratory analysis on the *3 disorders + healthy controls with NaN removed* dataset !

```python
import pandas as pd
df = pd.read_csv('/content/all_disorders_remove_std_onehot.csv')

# taking a peek
df.iloc[:, 1:].head()
```

```python
# features
features = list(df.columns[5:])
features
```

```python
# classes
classes = list(df.columns[1:5])
classes
```

```python
# correlation heatmap
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.figure(figsize=(18, 8))
matrix = np.triu(df[features].corr())
sns.heatmap(df[features].corr(), annot = True, cmap='RdPu', mask=matrix)
plt.title('Correlation Map Of Parkinson\'s Features')
```
